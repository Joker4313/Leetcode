# 排序算法
## 选择排序

```python
"""
开启一个循环，每轮从未排序区间选择最小的元素，将其放到已排序区间的末尾。
时间复杂度与数组初始顺序无关，为 O(N^2)
非稳定排序
"""


def selection_sort(nums: list[int]):
    n = len(nums)
    # 外循环：未排序区间为 [i, n-1]
    for i in range(n - 1):
        k = i  # k为最小值索引，假设当前索引的数最小，初始设为i
        # 内循环：找到未排序区间内的最小元素
        for j in range(i + 1, n):
            if nums[j] < nums[k]:
                k = j
        nums[i], nums[k] = nums[k], nums[i]


x = [5, 4, 3, 2, 1]
selection_sort(x)
print(x)
```

## 冒泡排序

```python
"""
连续地比较与交换相邻元素实现排序
时间复杂度与数组初始顺序有关，为 O(N^2)，如果某轮“冒泡”中没有执行任何交换操作，说明数组已经完成排序，可直接返回结果。
冒泡排序的最差时间复杂度和平均时间复杂度仍为 O(N^2) ；但当输入数组完全有序时，可达到最佳时间复杂度 O(N) 。
非稳定排序
"""


def bubble_sort(nums: list[int]):
    n = len(nums)
    # 反向遍历的原因：i 是「已排序区间的左边界」，天然代表了「下一轮内循环的终止位置」
    # 外循环：未排序区间为 [0, i]
    for i in range(n - 1, 0, -1):
        flag = False  # 判断是否发生了交换操作
        # 内循环：将未排序区间 [0, i] 中的最大元素交换至该区间的最右端
        for j in range(i):
            if nums[j] > nums[j + 1]:
                flag = True
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
        if not flag:
            break


x = [5, 4, 3, 2, 1]
bubble_sort(x)
print(x)
```

## 插入排序

```python
"""
在未排序区间选择一个基准元素，将该元素与其左侧已排序区间的元素逐一比较大小，并将该元素插入到正确的位置。
时间复杂度为为 O(N^2)，但在数据量较小的情况下，插入排序通常更快。因为单元计算操作数量小
"""


def insertion_sort(nums: list[int]):
    # 未排序区间[0, i-1]
    n = len(nums)
    for i in range(1, n):
        base = nums[i]
        j = i - 1
        # 内循环：将 base 插入到已排序区间 [0, i-1] 中的正确位置
        while j >= 0 and nums[j] > base:
            nums[j + 1] = nums[j]
            j -= 1
        nums[j + 1] = base


x = [5, 4, 3, 2, 1]
insertion_sort(x)
print(x)
```

# 快速排序

```python
def partition(nums: list[int], left: int, right: int) -> int:
    i, j = left, right
    base = nums[left]  # 以nums[left]为基准数
    while i < j:
        while i < j and nums[j] > base:
            j -= 1  # 从右往左查找第一个小于基准数的数
        while i < j and nums[i] < base:
            i += 1  # 从左往右查找第一个大于基准数的数
        nums[i], nums[j] = nums[j], nums[i]
    nums[left], nums[i] = nums[i], nums[left]

    return i


def quick_sort(nums: list[int], left: int, right: int):
    if left >= right:
        return
    pivot = partition(nums, left, right)
    quick_sort(nums, left, pivot - 1)
    quick_sort(nums, pivot + 1, right)


x = [5, 4, 3, 2, 1]
quick_sort(x, 0, len(x)-1)
print(x)
```
